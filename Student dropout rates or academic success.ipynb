```python
marital_status_map = {1: 'Single', 
                      2: 'Married', 
                      3: 'Widower', 
                      4: 'Divorced', 
                      5: 'Facto Union', 
                      6: 'Legally Separated'}

# group the data by marital status and calculate the mean grade for each group
marital_status_means = df.groupby('Marital status')[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].mean()

# create the bar chart
ax = marital_status_means.plot(kind='bar')
plt.xlabel('Marital Status')
plt.ylabel('Average Grade')
plt.title('Average Grade by Marital Status')

# set the x-axis labels
ax.set_xticklabels([marital_status_map[i] for i in range(1, 7)])
plt.show()
```

```python
nationality_map = {1: 'Portuguese', 2: 'German', 6: 'Spanish', 11: 'Italian', 13: 'Dutch', 14: 'English', 17: 'Lithuanian', 21: 'Angolan', 22: 'Cape Verdean', 24: 'Guinean', 25: 'Mozambican', 26: 'Santomean', 32: 'Turkish', 41: 'Brazilian', 62: 'Romanian', 100: 'Moldova (Republic of)', 101: 'Mexican', 103: 'Ukrainian', 105: 'Russian', 108: 'Cuban', 109: 'Colombian'}

# group the data by nationality and calculate the mean grade for each group
nationality_means = df.groupby('Nacionality')[['Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)']].mean()

# create the grouped bar chart
ax = nationality_means.plot(kind='bar')
plt.xlabel('Nationality')
plt.ylabel('Average Grade')
plt.title('Average Grade by Nationality')

# set the x-axis labels
ax.set_xticklabels([nationality_map[i] for i in nationality_means.index])
plt.show()
```
